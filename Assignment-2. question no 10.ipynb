```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
driver=webdriver.Chrome(r'/Users/rinki/Downloads/chromedriver')
```

```python
url='https://www.ambitionbox.com/'
driver.get(url)

#click on salary option

sal_opt=driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]')
sal_opt

sal_opt.click()

#click on search company/designation name
search_job=driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')
search_job

search_job.send_keys("Data Scientist")
```

```python
#click on desired option
find_job=driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')
find_job
find_job.click()
```

```python
#company name

comp_name=driver.find_elements_by_xpath('//div[@class="name"]')
comp_name

Company_name=[]
for i in comp_name:
    Company_name.append(i.text)
Company_name

#Average Salary

avg_sal=driver.find_elements_by_xpath('//div[@class="average-indicator-wrapper"]')
avg_sal

Avg_Salary=[]
for i in avg_sal:
    Avg_Salary.append(i.text)
Avg_Salary

#Minimum Salary

min_sal=driver.find_elements_by_xpath('//div[@class="value body-medium"]')
min_sal

Min_Salary=[]
for i in min_sal:
    Min_Salary.append(i.text)
Min_Salary[0:21:2]

#Maximum Salary

max_sal=driver.find_elements_by_xpath('//div[@class="value body-medium"]')
max_sal

Max_Salary=[]
for i in max_sal:
    Max_Salary.append(i.text)
Max_Salary[1:21:2]

#experience required

exp=driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
exp

Exp=[]
for i in exp:
    Exp.append(i.text.replace('Data Scientist\n . \n',''))
Exp

data=pd.DataFrame({})
data['Company_name']=Company_name
data['Avg_Salary']=Avg_Salary
data['Min_Salary']=Min_Salary[0:21:2]
data['Max_Salary']=Max_Salary[1:21:2]
data['Exp']=Exp

data
```
