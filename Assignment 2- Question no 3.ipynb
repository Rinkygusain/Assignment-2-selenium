```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

#open automated. chrome
driver=webdriver.Chrome(r'/Users/rinki/Downloads/chromedriver')
```

```python
#open the website
url='https://www.naukri.com/'
driver.get(url)

#search job skill,designation

search_title=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[1]/div/div/div/input')
search_title

#search "Data Analyst"
search_title.send_keys("Data Scientist")

#search the seleted options

search_button=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button
search_button.click()
```

```python
#scrape salary 3-6 lakhs

salary_check=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[2]/section[1]/div[2]/div[4]/div[2]/div[2]/label/i')
salary_check.click()
```

```python
#scrape job location

job_loc=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[2]/section[1]/div[2]/div[4]/div[2]/div[3]/label/i')
job_loc.click()
```

```python
#scarpe Job_title
job_title=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
job_title

Designation=[]
for i in job_title:
    Designation.append(i.text)
Designation

#Scrape location
job_location=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
job_location

Location=[]
for i in job_location:
    Location.append(i.text)
Location

#scrape company_name
org_detail=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
org_detail

company_name=[]
for i in org_detail:
    company_name.append(i.text)
company_name

#scrape experienced
exp_req=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
exp_req

Experience=[]
for i in exp_req:
    Experience.append(i.text)
Experience

#Make DataFrame of 10 Data Analyst
dataScientist=pd.DataFrame({})
dataScientist['Designation']=Designation[:10]
dataScientist['Location']=Location[:10]
dataScientist['company_name']=company_name[:10]
dataScientist['Experience']=Experience[:10]
dataScientist
```
